### Изменение типов данных в столбцах

Сначала заменим тип данных в столбце `user_score` на float64:

Посмотрим почему тип данных у столбца object, возможно в нем есть ошибки в виде строк и символов

```python
df['user_score'].unique()
```

Видим, что есть странная оценка "tbd". Посчитаем сколько таких строк

```python
#Посмотрим на игры со значением user_score равному "tbd"
df[df['user_score'] == 'tbd']
```

Всего видим 2424 игры с таким типом рейтинга

Аббревиатура TBD имеет несколько расшифровок. Нас же интересует игровая сфера. 'To Be Determined' - 'Будет определено'. Т.е. это то же самое, что неизвестное значение. Поэтому, можно смело заменить значение 'tbd' на 'NaN'.

```python
#Заменим tbd на NaN
df['user_score'] = df['user_score'].replace('tbd', np.nan, regex=True)
```

```python
# Поменяем формат столбца user_score на float
df['user_score'] = df['user_score'].astype('float64')
df['user_score'].dtype
```

Тип данных изменен корректно

Изменим тип данных в столбце `year_of_release` на Int64:

```python
# Изменим тип данных в столбце year_of_release 
df['year_of_release'] = df['year_of_release'].astype('Int64')
df['year_of_release'].dtypes
```

Тип данных изменен корректно

Можно попробовать восполнить пробелы в годах за счет наличия в датасете информации об одних и тех же играх, только выпущенных для разных платформ (в надежде, что у какой-то из версий игры будет запись о годе, как это оказалось с игрой "FIFA Soccer 2004").

## Исследовательский анализ данных

Посмотрим, сколько игр выпускалось в разные годы.

```python
# строим график динамики продаж игр по годам
df.pivot_table(index='year_of_release',
               values='total_sales',
               aggfunc='sum')\
  .sort_values(by='year_of_release')\
  .plot(kind='line',
        figsize=(10,4))
plt.xlabel('Год релиза')
plt.ylabel('Суммарные продажи, млн. копий')
plt.title('Динамика продаж игр по годам');
```

**Вывод**:
Из графика видно, что с 1995 года количество игр начинает ежегодно расти быстрыми темпами. Это связано с развитием компьютеров, появлением компакт-дисков и развитием игровой индустрии. Данный рост продолжается до 2008 года.

```python
def print_stat_corr(platforms):
    
    for name in platforms:
        platform = relevant_data[relevant_data['platform']==name]
        fig, ax = plt.subplots(1 ,2, figsize=(9,4))
        sns.scatterplot(x='user_score', y='total_sales', data=platform, ax=ax[0])
        sns.scatterplot(x='critic_score', y='total_sales', data=platform, ax=ax[1])
        fig.suptitle(name, fontsize=15)
        ax[0].set(xlabel='Оценка пользователей')
        ax[1].set(xlabel='Оценка критиков')
        ax[0].set(ylabel='Количество продаж')
        ax[1].set(ylabel='Количество продаж')
        plt.show()
    
        correl = platform['user_score'].corr(platform['total_sales'])
        print(f'Корреляция между отзывами пользователей и продажами: {correl}')
        critic_correl = platform['critic_score'].corr(platform['total_sales'])
        print(f'Корреляция между отзывами критиков и игровой платформой: {critic_correl}')
```

```python
print_stat_corr(list_of_top_6)
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**Сформулируем нулевую гипотезу H0:**

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

**Сформулируем альтернативную гипотезу H1:**

Средние пользовательские рейтинги платформ Xbox One и PC отличаются

**Рассмотрим распределения пользовательских рейтингов в зависимости от платформы**

Создадим выборки по пользовательким оценкам этих двух платформ

```python
xbox_rating = relevant_data.query('platform == "XOne"')['user_score']
pc_rating = relevant_data.query('platform == "PC"')['user_score']
```

Посмотрим на статистики по распределениям:

```python
xbox_rating.describe()
```

Среднее значение пользовательского рейтинга по играм на платформе XOne - 6.6, а стандартное отклонение - 1.4

```python
sns.histplot(xbox_rating, kde=True)
plt.xlabel('Пользовательский рейтинг')
plt.ylabel('Количество оценок');
```

Посчитаем дисперсию по выборке:

```python
np.var(xbox_rating)
```

Судя по дисперсии и стандартному отклонению разброс данных небольшой.

Теперь изучим те же статистики по платформе PC

```python
pc_rating.describe()
```

Среднее значение пользовательского рейтинга по играм на платформе PC - 6.3, а стандартное отклонение - 1.7

```python
sns.histplot(pc_rating, kde=True)
plt.xlabel('Пользовательский рейтинг')
plt.ylabel('Количество оценок');
```

Посчитаем дисперсию по выборке:

```python
np.var(pc_rating)
```

Дисперсия по пользовательским оценкам игр на платформе PC выше, чем на платформе XOne в силу большего разброса оценок

Для проверки гипотезы буду использовать Т-тест.

Зададим критический уровень статистической значимости:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
```

```python
results = st.ttest_ind(xbox_rating.dropna(),pc_rating.dropna(),equal_var=False)
```

Указываю параметр equal_var=False, поскольку дисперсия выборок разная

```python
display('p-значение:', results.pvalue)
if results.pvalue < alpha:
     print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод**:

Нулевую гипотезу отвергнуть нельзя, поскольку вероятность получить среднее значение пользовательской оценки игр на платформе PC  равным среднему значению пользовательской оценки игр на платфоре XOne равна 14%

Дисперсия по пользовательским оценкам игр жанра Sporst выше, чем жанра Action в силу большего разброса оценок

Для проверки гипотезы буду использовать Т-тест.

Зададим критический уровень статистической значимости:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
```

```python
results = st.ttest_ind(action_rating.dropna(),sports_rating.dropna(),equal_var=False)
```

Указываю параметр equal_var=False, поскольку дисперсия выборок разная

```python
display('p-значение:', results.pvalue)
if results.pvalue < alpha:
     print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

**Вывод:**
Получив p-value, мы отвергли Нулевую гипотезу. Таким образом, у нас практически нет вероятности получить одинакове средние рейнтинги по двум жанрам.
