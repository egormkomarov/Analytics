### Анализ пропущенных значений и артефактов

#### Исседование значений в стобце `ceiling_height`

Из первоначального анализа, знаем, что в столбце имеются пропуски. Необходимо изучить данные столбца, чтобы принять решение, на какие значения заменять пропуски.

```python
# статистика 
df['ceiling_height'].describe()
```

Судя по большому стандартному отклонению и максимальному значению высоты потолков в 100 м, в столбце явно имеются аномальные значения. Медианное значение в 2,65 м похоже на правду, но необходимо проверить какую долю составляют аномальные значения.\
Изучим распределения высоты потолков в заявках.

```python
# диаграмма размаха 
df.boxplot(column='ceiling_height')
```

Видим большое количество аномально больших значений высот потолков, а также присутствуют предельно низкие значения. Значение в 100 метров явно выбивается из общей картины.
Уменьшим масштаб,чтобы увидить границы допустимых значений.

```python
df.boxplot(column='ceiling_height')
plt.ylim(0,5)
```

Проанализируем объявления у которых высота потолков выше нормального значения (третий кватиль - 2,8 м + 1,5 межквартильного размаха) в 3,22 м. \
Согласно СНиПам, высота потолков в квартирах в среднем 2,7 м, но в Сталинских домах, высота потолков может достигать 4 метров. Поэтому за отсечку возьму именно эту максимально-возможную высоту.
Сгруппирую по локации и этажности дома. Возможно в завках представлены замки с высокими потолками.

```python
# выборка по аномально большим высотам потолков
df['celing_too_high'] = df['ceiling_height'] > 4
      
df.head()
```

```python
# считаем среднее значение высоты потолков из выборки и количество объявлений
df.query('celing_too_high == True') \
  .pivot_table(index = ['locality_name','floors_total'], 
               values='ceiling_height',
               aggfunc=['mean','count'],
               margins=True)
```

Такие данные выглядят сомнительно. Количество таких объявлений - 75 шт., что менее одного процента от общей выборки. Удалим их позже из датафрейма для уменьшения количества шумов и исключения дальнейшего негативного влияния на анализ данных.

Теперь рассмотрим строки с аномально низкими значениями высот потолков:

```python
# выборка аномально низких значений высот потолков
df['celing_too_low'] = df['ceiling_height'] < 2.2

df.head()
```

```python
# считаем среднее значение высоты потолков из выборки и количество объявлений
df.query('celing_too_low == True') \
  .pivot_table(index = ['locality_name','floors_total'], 
               values='ceiling_height',
               aggfunc=['mean','count'],
               margins=True)
```

Таких объявлений 14 штук, такие потолки не соответсвтуют СНиПам. Их также удалим позже из выборки.

```python
# удаляю из датафрейма объявления с аномально большими и аномально маленькики значениями высоты потолков
df = df.query('celing_too_low == False and celing_too_high == False')
```

```python
# новый размер датафрейма
df.shape
```

Произведем замену пустых значений в столбце на медианное значение, сгруппированное по локации и этажности дома:

```python
df['ceiling_height'] = df['ceiling_height'].fillna(
    df.groupby(['locality_name','floors_total'])['ceiling_height']
      .transform('median'))
```

```python
# проверка оставших пропусков в столбце
df['ceiling_height'].isna().sum()
```

Осталось 508 строк с пропусками в столбце. Заменим оставшиеся пропуски на медианное значение по этажности.

```python
# Заменим оставшиеся пропуски на медианное значение по этажности
df['ceiling_height'] = df['ceiling_height'].fillna(df.groupby('floors_total')['ceiling_height'].transform('median'))
```

Анализировать такие данные сложно, поскольку встречаются разные слова пос**е**лок и пос**ё**лок. Необходимо убрать этот артефакт. А также есть вариции "поселок городского типа" и "городской поселок". Приведем все к единообразию

### Расчет дополительных параметров

Проведем расчет **цены квадратного метра** в каждом объявлении:

```python
# расчет цены квадратного метра недвижимости
df['area_price'] = (df['last_price'] / df['total_area']).astype('int')
# проверка
df['area_price'].head()
```

Теперь посчитаем **соотношение площади кухни к общей площади**, а также **соотношение жилой площади к общей площади** недвижимости:

```python
# расчет соотношения площади кухни к общей площади
df['kitchen_ratio'] = round(df['kitchen_area'] / df['total_area'],2)
# проверка
df['kitchen_ratio'].head()
```

```python
# расчет соотношения площади кухни к общей площади
df['living_ratio'] = round(df['living_area'] / df['total_area'],2)
# проверка
df['living_ratio'].head()
```

Также необходимо **категоризировать этаж недвижимости** (первый, последний, другой):

```python
# категоризация данных
df['floor_name'] = np.where(df['floor'] == 1 ,'первый','другой')
df['floor_name'] = np.where((df['floor'] == 1) | (df['floor'] != df['floors_total']),df['floor_name'],'последний')
```

```python
# проверка
df[(df['floor_name'] == 'последний') & (df['floor'] == 1) & (df['floors_total'] == 1)]
```

```python
df[['floor_name','floor','floors_total']].head(10)
```

Данные категоризированы верно

Теперь добавим в таблицу **год, месяц и день недели размещения объявления**:

```python
# год публикации
df['year_exposition'] = pd.DatetimeIndex(df['first_day_exposition']).year
# месяц публикации
df['month_exposition'] = pd.DatetimeIndex(df['first_day_exposition']).month
# день недели публикации
df['week_day_exposition'] = pd.DatetimeIndex(df['first_day_exposition']).dayofweek
```

```python
def print_analitycs():
    analitycs = ['rooms','floor','city_centers_nearest_km','year_exposition']
    
    for feature in analitycs:   
        df_sp.query('center == "центр"')\
             .pivot_table(index=feature,
                          values='last_price')\
             .plot(kind='line',
                   grid=True,
                   title=f'Зависимость цены недвижимости от {feature}',
                   linewidth = 3)
```

```python
print_analitycs()
```

**Вывод:**
1) Цена недвижимости в центре Санкт-Петербурга напрямую зависит от количества комнат, аналогично как и по всей базе объявлений.\

3) Цена квартиры напрямую зависит от расположения от центра города, это применимо ко всем объявлениям.\
4) Цены на квартиры в центре Санкт-Петербурга падали в период с 2015 по 2017 гг. Аналогичная динамика была и по всем объявлениям в базе.
