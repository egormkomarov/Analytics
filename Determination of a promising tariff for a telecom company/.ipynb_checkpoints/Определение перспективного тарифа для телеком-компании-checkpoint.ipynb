## Анализ исходных данных

### Описание проекта

Клиентам федерального оператора сотовой связи «Мегалайн» предлагают два тарифных плана: **«Смарт»** и **«Ультра»**. 

Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

**Цель исследования**  —  сделать предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. \
Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

**Задачи исследования:**

1) Описать поведение клиентов оператора, исходя из выборки. \
2) Сколько минут разговора, сколько сообщений и какой объём интернет-трафика требуется пользователям каждого тарифа в месяц? \
3) Посчитать среднее количество, дисперсию и стандартное отклонение для метрик, построить гистограммы и описать распределения.\
4) Проверить следующие гипотезы:
* средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
* средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

5) Посчитать для каждого пользователя:
* количество сделанных звонков и израсходованных минут разговора по месяцам;
* количество отправленных сообщений по месяцам;
* объем израсходованного интернет-трафика по месяцам;
* помесячную выручку с каждого пользователя (вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану).

**Описание тарифов**

*Тариф «Смарт»*

Ежемесячная плата: 550 рублей\
Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика\
Стоимость услуг сверх тарифного пакета:\
минута разговора: 3 рубля\
сообщение: 3 рубля\
1 Гб интернет-трафика: 200 рублей

*Тариф «Ультра»*

Ежемесячная плата: 1950 рублей\
Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика\
Стоимость услуг сверх тарифного пакета:\
минута разговора: 1 рубль\
сообщение: 1 рубль\
1 Гб интернет-трафика: 150 рублей

**Примечание**

«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.

Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта. Примите, что неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.

```python
df_internet_users['sum_gb_used'] = np.ceil(df_internet_users['sum_mb_used'] / 1024).astype('int')
df_internet_users.head()
```

### Расчет помесячной выручки с каждого пользователя

```python
df_full.head()
```

```python
def revenue_calc(row):
    
    sum_gb_used = row['sum_gb_used']
    mb_per_month_included = row['mb_per_month_included']
    rub_per_gb = row['rub_per_gb']
    
    minute_duration = row['minute_duration']
    minutes_included = row['minutes_included']
    rub_per_minute = row['rub_per_minute']
    
    message_count = row['message_count']
    messages_included = row['messages_included']
    rub_per_message = row['rub_per_message']
    
    rub_monthly_fee = row['rub_monthly_fee']
    
    if sum_gb_used > (mb_per_month_included / 1024):
        revenue_internet = (sum_gb_used - mb_per_month_included / 1024) * rub_per_gb
    else:
        revenue_internet = 0
        
    if minute_duration > minutes_included:
        revenue_minutes = (minute_duration - minutes_included) * rub_per_minute
    else:
        revenue_minutes = 0
    
    if message_count > messages_included:
        revenue_message = (message_count - messages_included) * rub_per_message
    else:
        revenue_message = 0
    
    revenue = revenue_internet + revenue_minutes + revenue_message + rub_monthly_fee
    
    return revenue 
```

Тестируем функцию на тарифе smart

Базовый вариант - все условия тарифа без доплат выполнены

```python
row_values = [15,500,50,50,15360,500,550,200,3,3] #клиенту 36 лет, и у него есть работа
row_columns = ['sum_gb_used','minute_duration','message_count','messages_included','mb_per_month_included','minutes_included','rub_monthly_fee','rub_per_gb','rub_per_message','rub_per_minute']
row = pd.Series(data=row_values, index=row_columns) 
revenue_calc(row) 
```

Частный вариант - + 1 Гб к трафику

```python
row_values = [16,500,50,50,15360,500,550,200,3,3] #клиенту 36 лет, и у него есть работа
row_columns = ['sum_gb_used','minute_duration','message_count','messages_included','mb_per_month_included','minutes_included','rub_monthly_fee','rub_per_gb','rub_per_message','rub_per_minute']
row = pd.Series(data=row_values, index=row_columns) 
revenue_calc(row) 
```

Частный вариант - + 1 Гб к трафику, + 1 минут сверх пакета

```python
row_values = [16,501,50,50,15360,500,550,200,3,3] #клиенту 36 лет, и у него есть работа
row_columns = ['sum_gb_used','minute_duration','message_count','messages_included','mb_per_month_included','minutes_included','rub_monthly_fee','rub_per_gb','rub_per_message','rub_per_minute']
row = pd.Series(data=row_values, index=row_columns) 
revenue_calc(row) 
```

## Проверка гипотез

Необходимо проверить следующие гипотезы:

- средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
- средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионо

### Гипотеза 1: Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

**Сформулируем нулевую гипотезу H0:**\
Средняя выручка пользователей тарифов «Ультра» и «Смарт» одинакова

Т.е средняя выручка пользователей тарифа "Ультра" равна средней выручке пользователей тарифа "Смарт"

**Альтернативная гипотеза H1:** \
Средняя выручка пользователей тарифа "Ультра" не равна средней выручке пользователей тарифа "Смарт"

Посмотрим на распределение средней выручки с пользователя в зависимости от вида тарифа

```python
sns.histplot(x='revenue',
             hue='tariff_name',
             bins=20,
             data=df_full)\
   .set(xlabel='Средняя выручка с пользователя',
        ylabel='Количество пользователей');
```

Видим, что распределения по обоим тарифам не имеют нормальный характер распределения.

Посчитаем дисперсии этих распределений:

Дисперсия по выручке на пользователя в тарифе smart:

```python
np.var(df_full.query('tariff_name == "smart"')['revenue'])
```

Дисперсия по выручке на пользователя в тарифе ultra:

```python
np.var(df_full.query('tariff_name == "ultra"')['revenue'])
```

Видим большу разницу в дисперсии распредлений этих выборок.

Для проверки гипотезы буду использовать Т-тест.

Необходимо подготовить две выборки по выручке для каждого из тарифов.

```python
revenue_smart = df_full.query('tariff_name == "smart"')['revenue']
revenue_ultra = df_full.query('tariff_name == "ultra"')['revenue']
```

Зададим критический уровень статистической значимости:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
```

```python
results = st.ttest_ind(revenue_smart, revenue_ultra, equal_var=False)
```

Указываю параметр equal_var=False, поскольку выборки у нас разных размеров и дисперсия по ним разная

```python
display('p-значение:', results.pvalue)
```

```python
if results.pvalue < alpha:
     print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевую гипотезу нужно отвергнуть, поскольку вероятность получить такое же значение средней выручки на пользователя в тарифе smart, как в тарифе ultra, ничтожно мала.

### Гипотеза 2: Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

**Сформулируем нулевую гипотезу H0:**\
Средняя выручка пользователей из Москвы равна выручке пользователей из других регионов

**Альтернативная гипотеза H1:** \
Средняя выручка пользователей из Москвы не равна выручке пользователей из других регионов

Рассмотрим распредления выручки по пользователям из Москвы и по пользователям из других регионов

```python
revenue_Moscow = df_full.query('city == "Москва"')['revenue']
revenue_other = df_full.query('city != "Москва"')['revenue']
```

Статистика по выручке на пользователя в других регионах:

```python
sns.histplot(bins=20,
             data=revenue_other)\
   .set(xlabel='Средняя выручка с пользователя в других регионах',
        ylabel='Количество пользователей');
```

Видим, что распределения по обоим тарифам не имеют нормальный характер распределения.

Посчитаем дисперсии этих распределений:

Дисперсия по выручке на пользователя в Москве:

```python
np.var(revenue_Moscow)
```

Дисперсия по выручке на пользователя в других регинах:

```python
np.var(revenue_other)
```

Видим большу разницу в дисперсии распредлений этих выборок.

Для проверки гипотезы буду использовать Т-тест.

Зададим критический уровень статистической значимости:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
```

```python
results = st.ttest_ind(revenue_Moscow,revenue_other,equal_var=False)
```

Указываю параметр equal_var=False, поскольку выборки у нас разных размеров и дисперсия по ним разная

```python
display('p-значение:', results.pvalue)
```

```python
if results.pvalue < alpha:
     print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевую гипотезу отвергнуть нельзя, поскольку вероятность получить среднее значение выручки на пользователя в Москве равное среднему значению выручки на пользователя в других регионах равна 78%.
